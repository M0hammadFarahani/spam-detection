```python
df = pd.read_csv('/content/gdrive/MyDrive/enron_spam_data.csv')
```

```python
df.head()
```

```python
df = df[~df['Message'].isna()]
```

```python
label_list = list()

for label in df['Spam/Ham']:
  if label == 'spam':
    label_list.append(1)
  elif label == 'ham':
    label_list.append(0)

print(label_list)
```

```python
text_list = list()

for index in range(len(df)):
  text = str(df.iloc[index]['Subject']) + str(df.iloc[index]['Message'])
  text_list.append(text)

print(text_list[:5])
```

```python
count_dict = dict()

for index, text in enumerate(text_list):
  if label_list[index] == 1:

    splited_text = text.split()

    for word in splited_text:
      if len(word) >= 3:
        if word in count_dict:
          count_dict[word] += 1
        else:
          count_dict[word] = 1

print(count_dict)
```

```python
sorted_count_dict = sorted(count_dict.items(), key= lambda item: item[1], reverse=True)[:500]
```

```python
spam_words = list()

for word, count in sorted_count_dict:
  spam_words.append(word)

print(spam_words)
```

```python
def check_spam_in_text(email, spam_words):
  vector = np.zeros((1, len(spam_words)), dtype=np.int32)

  splited_email = email.split()

  for index, word in enumerate(spam_words):
    if word in splited_email:
      vector[0, index] = 1

  return vector
```

```python
sample = check_spam_in_text(text_list[21410], spam_words)

print(sample)
```

```python
preprocessed_emails = np.empty((len(text_list), len(spam_words)), dtype=np.int32)

for index, email in enumerate(text_list):
  feature_vector = check_spam_in_text(email, spam_words)
  preprocessed_emails[index, :] = feature_vector
```

```python
print(preprocessed_emails[:10])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(preprocessed_emails, label_list,
                                                    test_size=0.2, shuffle=True, stratify=label_list)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(f'Train Accuracy: {accuracy_score(train_preds, y_train):.2f}')
print(f'Test Accuracy: {accuracy_score(test_preds, y_test):.2f}')
```
